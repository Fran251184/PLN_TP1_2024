# FUNCIONES DE PRE-PROCESAMIENTO DE TEXTO Y DATOS

```python
import re
import unicodedata

def limpiar_y_normalizar_texto(texto):
    """
    Limpia y normaliza una cadena de texto.

    - Convierte todas las letras a minúsculas.
    - Elimina caracteres especiales excepto espacios.
    - Maneja adecuadamente acentos y diéresis.
    - No elimina palabras de un solo carácter.

    Args:
    texto (str): La cadena de texto a limpiar y normalizar.

    Returns:
    str: La cadena de texto limpia y normalizada.
    """
    # Convertir a minúsculas
    texto = texto.lower()

    # Normalizar acentos y diéresis
    texto = unicodedata.normalize('NFD', texto)
    texto = texto.encode('ascii', 'ignore').decode('utf-8')

    # Eliminar caracteres especiales (solo dejar letras y espacios)
    texto = re.sub(r'[^a-z\s]', '', texto)

    # Eliminar espacios adicionales
    texto = re.sub(r'\s+', ' ', texto).strip()

    return texto


def transformar_str_en_lista(texto):
    """
    Transforma una cadena de texto separada por comas en una lista de elementos.

    Args:
    texto (str): La cadena de texto a transformar.

    Returns:
    list: Lista de elementos separados por comas.
    """
    return [item.strip() for item in texto.split(',')]
```

# FUNCIONES DE PROCESAMIENTO DE TEXTO PARA BÚSQUEDA SEMÁNTICA (EMBEDDINGS) O MEDIANTE DISTANCIA (LEVENSHTEIN)

```python
import Levenshtein as lev

def buscar_levenshtein(df, string, nombre_columna):
    """
    Busca registros en el DataFrame que coincidan con el string proporcionado utilizando la distancia de Levenshtein.

    Args:
    df (pd.DataFrame): DataFrame que contiene los datos.
    string (str): El string a buscar.
    nombre_columna (str): El nombre de la columna que contiene las listas de strings.
    umbral (int): El umbral de distancia de Levenshtein para considerar coincidencias.

    Returns:
    pd.DataFrame: DataFrame filtrado con los registros que tienen al menos un elemento en la columna especificada
    que sea similar al string proporcionado dentro del umbral de distancia de Levenshtein.
    """
    umbral = 5

    def matchea(lista):
        for i in lista:
            if lev.distance(string, limpiar_y_normalizar_texto(i)) <= umbral:
                return True
        return False

    # Filtrar los registros con la condición de distancia de Levenshtein
    df_filtrado = df[df[nombre_columna].apply(matchea)]

    return df_filtrado

from sentence_transformers import SentenceTransformer

def descripcion_transformer(descripcion):
    """
    Genera un embedding para una descripción utilizando un modelo multilingüe de Sentence Transformers.

    Parámetros:
    descripcion (str): La descripción del libro (o cualquier texto) para la cual se generará el embedding.

    Retorna:
    numpy.ndarray: El embedding de la descripción.
    """
    # Cargar un modelo multilingüe
    model = SentenceTransformer('distiluse-base-multilingual-cased-v1')

    # Generar el embedding para la descripción
    descripcion_embedding = model.encode(descripcion)

    return descripcion_embedding

import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def busca_libro_cercania(vector_descripcion, df):
    """
    Busca los tres libros más similares a una descripción dada utilizando la similitud del coseno.

    Args:
    vector_descripcion (numpy.ndarray): Vector de características que representa la descripción ingresada.
    df (pd.DataFrame): DataFrame que contiene los libros, incluyendo una columna con los embebidos de las sinopsis.

    Returns:
    list: Una lista de diccionarios, cada uno representando un libro. Cada diccionario contiene el título, los autores,
    los géneros y la sinopsis del libro.
    """
    # Convertir vector_descripcion a un array de 2D si no lo es
    if vector_descripcion.ndim == 1:
        vector_descripcion = vector_descripcion.reshape(1, -1)

    # Obtener la matriz de embebidos de las sinopsis
    synopsis_matrix = np.stack(df["Embebbed_Synopsis"].values)

    # Calcular la similitud del coseno
    similarity_matrix = cosine_similarity(vector_descripcion, synopsis_matrix)

    # Encontrar los índices de las tres frases más semejantes
    most_similar_indices = similarity_matrix[0].argsort()[-3:][::-1]

    # Crear el diccionario con los resultados
    resultados = []
    for idx in most_similar_indices:
        resultado = {
            "Title": df.iloc[idx]["Title"],
            "Authors": df.iloc[idx]["Authors"],
            "Genres": df.iloc[idx]["Genres"],
            "Synopsis": df.iloc[idx]["Synopsis"]
        }
        resultados.append(resultado)

    return resultados
```
